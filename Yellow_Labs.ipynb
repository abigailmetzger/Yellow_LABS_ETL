```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url = 'https://www.akc.org/most-popular-breeds/2018-full-list/'
```

```python
response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
```

```python
results = soup.find_all('td')
# print(results[1:5])
```

```python
breeds = []
ranks = []
counter = 0
for result in results[2:]:
    data = result.text
    if (counter % 2) == 0:
        breeds.append(data)
    else:
        ranks.append(data)
    counter += 1
#     print(data)
#     print("------")
```

```python
# test_df = pd.DataFrame(list(zip(breeds, ranks)))
# test_df.head()
```

```python
rank_df = pd.DataFrame({"Breeds": breeds, "Ranks": ranks})
rank_df.sort_values("Breeds").head(10)
```
